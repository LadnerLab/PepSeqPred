```python
from pathlib import Path
import pandas as pd

tsv_path = Path("../data/input_data_20_4_10_all.tsv")
exclude_path = Path("./peptidePredictionTestingList.txt")
```

```python
exclude_codes = set(line.strip()
                    for line in exclude_path.read_text().splitlines() 
                    if line.strip())
len(exclude_codes)
```

```python
df = pd.read_csv(tsv_path, sep='\t')
filtered_df = df[~df["CodeName"].isin(exclude_codes)]
filtered_df.head()
```

```python
filtered_df.to_csv("../data/input_data_20_4_10_all_rm_special.tsv", sep="\t", index=False)
```
