```python
from pathlib import Path
import csv
```

```python
emb_dir = Path("../localdata/sample/embeddings")
meta_in = Path("../localdata/input_data_20_4_10_all.tsv")
meta_out = Path("../localdata/sample/sample_meta.tsv")
```

```python
emb_ids = {p.stem for p in emb_dir.glob("*.pt")}
len(emb_ids)
```

```python
rows_out = 0
proteins_out = set()

with meta_in.open("r", newline="", encoding="utf-8") as fin, \
     meta_out.open("w", newline="", encoding="utf-8") as fout:
    r = csv.DictReader(fin, delimiter="\t")
    w = csv.DictWriter(fout, fieldnames=r.fieldnames, delimiter="\t", lineterminator="\n")
    w.writeheader()

    for row in r:
        fn = (row.get("FullName") or "").strip()
        pid = fn.split()[0][3:] if fn.startswith("ID=") else ""
        if pid in emb_ids:
            w.writerow(row)
            rows_out += 1
            proteins_out.add(pid)
```

```python
print(f"Wrote {meta_out}")
print(f"rows: {rows_out}")
print(f"unique proteins: {len(proteins_out)}")
```
