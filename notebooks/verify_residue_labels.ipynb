```python
import torch
from pathlib import Path
```

```python
labels_path = Path("../data/sample/labels.pt")
obj = torch.load(labels_path)
print(f"Keys: {obj.keys()}")
```

```python
labels = obj["labels"]
proteins = obj["proteins"]
```

```python
# inspect head of labels
count = 0
for k, v in labels.items():
    print(f"{k} ---> {v}")
    count += 1
    if count > 4: break
```

```python
# inspect head of proteins
count = 0
for k, v in proteins.items():
    print(f"{k} ---> {v}")
    count += 1
    if count > 4: break
```
